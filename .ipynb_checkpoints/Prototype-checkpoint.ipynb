## This is a test with one classifier per column, it does not improve the accuracy and is harder to analyze, so I decided to keep it simpler.

```python
import pandas as pd

vanilla_df = pd.read_csv('data.csv', sep=';', index_col=0)

vanilla_df.head(20)
```

```python
# Check how often Buchung and Valuta date are actually different
count = 0
for i in range(len(vanilla_df)):
    if vanilla_df['Buchungstag'][i]!=vanilla_df['Valutadatum'][i]:
        count += 1
count
```

```python
import math

# Conclusion of inital data analysis: drop Wahrung and Valutadatum as they are of no real use
reduced_df = vanilla_df.drop(columns=['Waehrung', 'Valutadatum'])

# Condense Buchungstag date to day only as month and year seldomly change
reduced_df['Buchungstag'] = reduced_df['Buchungstag'].apply(lambda x: x[0:2])

# Insert column Abbuchung which is true for negative transactions
reduced_df.insert(len(reduced_df.columns)-1, 'Abbuchung', reduced_df['Betrag'].apply(lambda x: float(x.replace(',', '.')) < 0))

# Remove sign from Betrag values and round them to nearest ten as it is more about coarse amount than exact values
reduced_df['Betrag'] = reduced_df['Betrag'].apply(lambda x: math.ceil(abs(float(x.replace(',', '.')))/10.0)*10)

# For each column except labels: convert to string, remove case and punctuation/special characters
for col in reduced_df.columns[:-1]:
    reduced_df[col] = reduced_df[col].apply(str).str.replace('\W', ' ').str.lower()

reduced_df.head(20)
```

```python
# split data
randomized_df = reduced_df.sample(frac=1)#, random_state=2)

split_index = round(len(randomized_df) * 0.75)

training_df = randomized_df[:split_index]
test_df = randomized_df[split_index:]

print(training_df.shape)
print(test_df.shape)
```

```python
from sklearn.feature_extraction.text import CountVectorizer
from sklearn.naive_bayes import GaussianNB

labels = training_df["label"].array
cvecs = list()
#feature_counts = list()
models = list()

for i, col in enumerate(training_df.columns[:-1]):
    cvec = CountVectorizer()
    cvec.fit(list(training_df[col]))
    counts = cvec.transform(list(training_df[col])).toarray()
    model = GaussianNB()
    model.fit(counts, labels)
    
    cvecs.append(cvec)
    #feature_counts.append(counts)
    models.append(model)
    
```

```python
test_labels = test_df["label"].array
classes = models[0].classes_

def classify_test_df_at(index):
    
    prob_sum = list(0 for i in range(len(classes)))
    sample = test_df.iloc[index]
    
    for i, col in enumerate(test_df.columns[:-1]):
        counts = cvecs[i].transform(sample[col].split()).toarray()
        probs = models[i].predict_log_proba(counts)
        prob_sum = [sum(x) for x in zip(prob_sum, probs[0])]
        
    return classes[prob_sum.index(max(prob_sum))]

# Test model
right = 0
wrong = 0
for i in range(len(test_df)):
    actual = test_labels[i]
    prediction = classify_test_df_at(i)
    print(str(i) + ") Prediction: " + prediction + " | Actual: " + actual)
    if prediction == actual:
        right += 1
    else:
        wrong += 1

print(str(right) + " correct predictions")
print(str(wrong) + " failed predictions")
print(str(right/len(test_df)*100) + "% success rate")
```
